```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Download market data
print("Downloading market data...")

# Brent Crude Oil, GBP/USD, FTSE 100, S&P 500
tickers = {
    'Brent_Crude': 'BZ=F',
    'GBPUSD': 'GBPUSD=X',
    'FTSE100': '^FTSE',
    'SP500': '^GSPC'
}

data = {}
for name, ticker in tickers.items():
    df = yf.download(ticker, start='2025-01-01', end='2026-04-16', progress=False)
    data[name] = df['Close']
    print(f"{name} downloaded — {len(df)} trading days")

print("\nAll data downloaded successfully")
```

```python
# Combine all data into one dataframe
prices = pd.concat(data, axis=1)
prices.columns = list(tickers.keys())
prices.index = pd.to_datetime(prices.index)

# Drop any rows where all values are missing
prices = prices.dropna(how='all')

# Normalise to 100 at start so we can compare everything on same chart
normalised = prices / prices.iloc[0] * 100

print("Data shape:", prices.shape)
print("\nFirst few rows:")
print(prices.head())
print("\nLatest prices:")
print(prices.tail())
```

## Data Cleaning

Financial market data pulled from APIs rarely arrives in perfect condition. Different markets operate on different trading calendars, the FTSE 100 closes on UK bank holidays while US markets remain open, and vice versa. This creates gaps in our dataset that need to be handled carefully before any analysis can be conducted.

**Missing values identified:**
- Brent Crude: 9 missing values (2.71%)
- FTSE 100: 7 missing values (2.11%)
- S&P 500: 11 missing values (3.31%)
- GBP/USD: 1 missing value (0.30%)

All assets fell below the 5% threshold considered acceptable in professional data analysis. The missing values are attributable to market holiday mismatches rather than data quality issues.

**Cleaning approach — forward fill:**
Missing values were filled using the previous trading day's closing price. This is standard practice in financial data analysis because asset prices do not change on days when markets are closed. A forward fill accurately reflects the economic reality that the last known price remains the best estimate of value on a non-trading day.

A backward fill was applied subsequently to handle any gaps at the very start of the dataset where no prior value exists to forward fill from.

The raw data was preserved separately throughout, ensuring the original downloaded prices remain intact for reference.

```python
# Check for missing values in each column
print("Missing values per asset:")
print(prices.isnull().sum())

print("\nPercentage missing:")
print((prices.isnull().sum() / len(prices) * 100).round(2))
```

```python
# Forward fill missing values
# Uses the previous trading day's price to fill gaps
prices_clean = prices.ffill()

# Backward fill any remaining gaps at the start of the dataset
prices_clean = prices_clean.bfill()

# Confirm no missing values remain
print("Missing values after cleaning:")
print(prices_clean.isnull().sum())

print("\nData is clean and ready for analysis")
```

## Analysis Framework

To answer the central question of whether geopolitical shocks create systemic or isolated market risk, five analytical techniques were applied to the data:

**Daily Returns**
Rather than analysing raw prices, daily percentage returns were calculated for each asset. This allows us to measure how much each asset moved from one trading day to the next, identifying unusually large or volatile movements that signal market stress. Returns also allow meaningful comparison between assets that trade at very different price levels.

**Price Normalisation**
All four assets were normalised to a base of 100 at January 2025. This removes the effect of different price scales and allows direct visual comparison of how each asset moved relative to its starting point over the full period. An asset reading 130 has gained 30% from its starting price regardless of whether it trades at $70 or $10,000.

**30-Day Rolling Volatility**
Volatility was calculated as the standard deviation of daily returns over a rolling 30-day window, annualised by multiplying by the square root of 252 — the number of trading days in a year. This shows how uncertainty evolved over time and identifies periods of elevated market stress around specific geopolitical events.

**Correlation Matrix**
A Pearson correlation matrix was constructed from daily returns across all four assets. This measures the strength and direction of relationships between assets, when one falls, does another tend to fall simultaneously? Importantly, correlation identifies co-movement but does not imply causation. Where assets move together it is likely because they share a 
common driver such as geopolitical risk, rather than one directly causing the other to move.

**Event Study Analysis**
For each of the six key geopolitical events, the 5-day return immediately before and after the event date was calculated. The difference between these, the shock impact, quantifies how much each event moved markets beyond their pre-event trajectory. This is standard methodology in financial economics for isolating the market impact of specific events.

**Real World Impact Quantification**
To translate the market data into tangible terms, the financial impact of the March 2026 Iran conflict was calculated for two real world scenarios,a UK business importing $1,000,000 of goods and an institutional investor holding a £10,000,000 FTSE 100 portfolio. This grounds the analysis in practical consequences rather than abstract percentages, demonstrating how geopolitical shocks translate directly into business costs and portfolio losses.

```python
# Calculate daily percentage returns
returns = prices_clean.pct_change().dropna()
returns = returns * 100  # Convert to percentage

print("=" * 60)
print("DESCRIPTIVE STATISTICS — DAILY RETURNS (%)")
print("=" * 60)

ret_desc = returns.describe().round(4)
ret_desc.loc['skewness'] = returns.skew().round(4)
ret_desc.loc['kurtosis'] = returns.kurtosis().round(4)

print(ret_desc)

print("\n")
print("=" * 60)
print("WHAT THIS TELLS US")
print("=" * 60)
print(f"Average daily oil move: {ret_desc.loc['mean', 'Brent_Crude']:.3f}%")
print(f"Largest single day oil gain: {ret_desc.loc['max', 'Brent_Crude']:.2f}%")
print(f"Largest single day oil drop: {ret_desc.loc['min', 'Brent_Crude']:.2f}%")
print(f"\nAverage daily GBP/USD move: {ret_desc.loc['mean', 'GBPUSD']:.3f}%")
print(f"Average daily FTSE move: {ret_desc.loc['mean', 'FTSE100']:.3f}%")
```

```python
# Define key geopolitical events
events = {
    'Trump Tariffs\nAnnounced': '2025-04-02',
    'Israel Strikes\nIran': '2025-06-13',
    'Iran Ceasefire': '2025-06-25',
    'OPEC+ Output\nIncrease': '2025-06-01',
    'US/Israel\nStrikes Iran': '2026-03-01',
    'Oil Peaks\n$118': '2026-03-15',
}

# Plot normalised prices with events
fig, ax = plt.subplots(figsize=(16, 8))

colors = ['#e67e22', '#2ecc71', '#3498db', '#e74c3c']
labels = ['Brent Crude Oil', 'GBP/USD', 'FTSE 100', 'S&P 500']

for col, color, label in zip(normalised.columns, colors, labels):
    ax.plot(normalised.index, normalised[col], 
            color=color, linewidth=2, label=label)

# Add event lines
for event_name, event_date in events.items():
    date = pd.to_datetime(event_date)
    ax.axvline(x=date, color='black', linestyle='--', 
               linewidth=1, alpha=0.7)
    ax.text(date, ax.get_ylim()[1] * 0.98, event_name,
            rotation=90, fontsize=7, va='top', ha='right')

# Formatting
ax.set_title('Geopolitical Shocks and Financial Market Volatility 2025-2026\nNormalised to 100 at January 2025', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Normalised Price (Base = 100)', fontsize=12)
ax.legend(loc='upper left', fontsize=10)
ax.grid(True, alpha=0.3)
ax.axhline(y=100, color='grey', linestyle='-', linewidth=0.5, alpha=0.5)

plt.tight_layout()
plt.savefig('geopolitical_market_volatility.png', dpi=300, bbox_inches='tight')
plt.show()

print("Chart saved as geopolitical_market_volatility.png")
```

```python
# Rolling 30-day volatility (annualised)
rolling_vol = returns.rolling(window=30).std() * np.sqrt(252)

fig, ax = plt.subplots(figsize=(16, 8))

colors = ['#e67e22', '#2ecc71', '#3498db', '#e74c3c']
labels = ['Brent Crude Oil', 'GBP/USD', 'FTSE 100', 'S&P 500']

for col, color, label in zip(rolling_vol.columns, colors, labels):
    ax.plot(rolling_vol.index, rolling_vol[col],
            color=color, linewidth=2, label=label)

# Add event lines
for event_name, event_date in events.items():
    date = pd.to_datetime(event_date)
    ax.axvline(x=date, color='black', linestyle='--',
               linewidth=1, alpha=0.7)
    ax.text(date, ax.get_ylim()[1] * 0.98, event_name,
            rotation=90, fontsize=7, va='top', ha='right')

ax.set_title('30-Day Rolling Volatility by Asset Class 2025-2026\n(Annualised)', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Annualised Volatility (%)', fontsize=12)
ax.legend(loc='upper left', fontsize=10)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('rolling_volatility.png', dpi=300, bbox_inches='tight')
plt.show()

print("Volatility chart saved")
```

```python
# Correlation matrix of daily returns
correlation = returns.corr()

fig, ax = plt.subplots(figsize=(10, 8))

# Create heatmap
mask = np.zeros_like(correlation, dtype=bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(correlation, 
            annot=True, 
            fmt='.3f',
            cmap='RdYlGn',
            center=0,
            vmin=-1, 
            vmax=1,
            ax=ax,
            linewidths=0.5,
            annot_kws={'size': 14, 'weight': 'bold'})

ax.set_title('Correlation Matrix of Daily Returns 2025-2026\nGeopolitical Risk Period', 
             fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

print("Correlation matrix saved")
```

```python
# Event Study Analysis
# Measure market impact in 5 days before and after each event

event_study_results = []

for event_name, event_date in events.items():
    event_dt = pd.to_datetime(event_date)
    
    # Find the nearest trading day
    idx = prices_clean.index.searchsorted(event_dt)
    if idx >= len(prices_clean):
        continue
    
    # Get 5 days before and after
    pre_start = max(0, idx - 5)
    post_end = min(len(prices_clean), idx + 6)
    
    pre_period = prices_clean.iloc[pre_start:idx]
    post_period = prices_clean.iloc[idx:post_end]
    
    if len(pre_period) == 0 or len(post_period) == 0:
        continue
    
    for asset in prices_clean.columns:
        pre_return = ((pre_period[asset].iloc[-1] - pre_period[asset].iloc[0]) 
                     / pre_period[asset].iloc[0] * 100)
        post_return = ((post_period[asset].iloc[-1] - post_period[asset].iloc[0]) 
                      / post_period[asset].iloc[0] * 100)
        
        event_study_results.append({
            'Event': event_name,
            'Date': event_date,
            'Asset': asset,
            'Pre_5day_Return': round(pre_return, 3),
            'Post_5day_Return': round(post_return, 3),
            'Shock_Impact': round(post_return - pre_return, 3)
        })

event_df = pd.DataFrame(event_study_results)

print("=" * 70)
print("EVENT STUDY — 5 DAY PRE AND POST RETURNS AROUND KEY GEOPOLITICAL EVENTS")
print("=" * 70)

for event in event_df['Event'].unique():
    print(f"\n{event} ({event_df[event_df['Event']==event]['Date'].iloc[0]})")
    print("-" * 50)
    subset = event_df[event_df['Event']==event][['Asset','Pre_5day_Return','Post_5day_Return','Shock_Impact']]
    print(subset.to_string(index=False))

print("\n")
print("=" * 70)
print("LARGEST SHOCKS ACROSS ALL EVENTS")
print("=" * 70)
print(event_df.nlargest(5, 'Shock_Impact')[['Event','Asset','Pre_5day_Return','Post_5day_Return','Shock_Impact']].to_string(index=False))
print("\nLARGEST NEGATIVE SHOCKS:")
print(event_df.nsmallest(5, 'Shock_Impact')[['Event','Asset','Pre_5day_Return','Post_5day_Return','Shock_Impact']].to_string(index=False))
```

```python
# Event Study Visualisation
fig, ax = plt.subplots(figsize=(16, 8))

# Pivot for grouped bar chart
pivot_df = event_df.pivot(index='Event', columns='Asset', values='Shock_Impact')

# Clean event names for display
pivot_df.index = [e.replace('\n', ' ') for e in pivot_df.index]

x = np.arange(len(pivot_df.index))
width = 0.2
colors = ['#e67e22', '#2ecc71', '#3498db', '#e74c3c']

for i, (asset, color) in enumerate(zip(pivot_df.columns, colors)):
    bars = ax.bar(x + i * width, pivot_df[asset], 
                  width, label=asset, color=color, alpha=0.85)

ax.set_title('Event Study — 5 Day Shock Impact by Asset Class\nKey Geopolitical Events 2025-2026',
             fontsize=14, fontweight='bold', pad=20)
ax.set_xlabel('Event', fontsize=12)
ax.set_ylabel('Shock Impact (%)', fontsize=12)
ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(pivot_df.index, rotation=15, ha='right', fontsize=9)
ax.legend(fontsize=10)
ax.axhline(y=0, color='black', linewidth=0.8)
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('event_study.png', dpi=300, bbox_inches='tight')
plt.show()

print("Event study chart saved")
```

```python
# Find nearest trading days to our event dates
def get_nearest_trading_day(date_str, df):
    date = pd.to_datetime(date_str)
    idx = df.index.searchsorted(date)
    idx = min(idx, len(df.index) - 1)
    return df.index[idx]

# Get actual trading days
date_pre = get_nearest_trading_day('2026-03-01', prices_clean)
date_post = get_nearest_trading_day('2026-03-15', prices_clean)

print(f"Using dates: {date_pre.date()} to {date_post.date()}")

# Real World Business Impact Analysis
print("\n" + "=" * 65)
print("REAL WORLD IMPACT — UK BUSINESS IMPORTING $1,000,000 OF GOODS")
print("=" * 65)

gbpusd_pre = prices_clean.loc[date_pre, 'GBPUSD']
gbpusd_post = prices_clean.loc[date_post, 'GBPUSD']

oil_pre = prices_clean.loc[date_pre, 'Brent_Crude']
oil_post = prices_clean.loc[date_post, 'Brent_Crude']

usd_amount = 1_000_000
gbp_cost_pre = usd_amount / gbpusd_pre
gbp_cost_post = usd_amount / gbpusd_post
gbp_difference = gbp_cost_post - gbp_cost_pre

print(f"\nGBP/USD before strikes: {gbpusd_pre:.4f}")
print(f"GBP/USD after strikes:  {gbpusd_post:.4f}")
print(f"\nCost of $1,000,000 in GBP before: £{gbp_cost_pre:,.0f}")
print(f"Cost of $1,000,000 in GBP after:  £{gbp_cost_post:,.0f}")
print(f"Additional cost to UK importer:   £{gbp_difference:,.0f}")

print(f"\nOil price before strikes: ${oil_pre:.2f}/barrel")
print(f"Oil price after strikes:  ${oil_post:.2f}/barrel")
print(f"Oil price increase:       {((oil_post-oil_pre)/oil_pre*100):.1f}%")

barrels_per_container = 1800
shipping_containers = 10
total_barrels = barrels_per_container * shipping_containers
fuel_cost_increase = (oil_post - oil_pre) * total_barrels * 0.3

print(f"\nEstimated additional fuel cost for 10 shipping containers:")
print(f"£{fuel_cost_increase:,.0f}")

print("\n" + "=" * 65)
print("INVESTOR PORTFOLIO IMPACT")
print("=" * 65)

portfolio_size = 1_000_000
ftse_pre = prices_clean.loc[date_pre, 'FTSE100']
ftse_post = prices_clean.loc[date_post, 'FTSE100']
ftse_return = (ftse_post - ftse_pre) / ftse_pre * 100
portfolio_impact = portfolio_size * (ftse_return / 100)

print(f"\nFTSE 100 before strikes: {ftse_pre:.0f}")
print(f"FTSE 100 after strikes:  {ftse_post:.0f}")
print(f"FTSE return over period: {ftse_return:.2f}%")
print(f"\nImpact on £1,000,000 FTSE portfolio: £{portfolio_impact:,.0f}")
print(f"Impact on £10,000,000 FTSE portfolio: £{portfolio_impact*10:,.0f}")
```

### Rolling Volatility

The rolling volatility chart shows not just what happened but **when markets were most uncertain and afraid**. Oil volatility dominated throughout, consistently running at two to four times the level of equity and currency volatility, confirming oil as the primary channel for geopolitical risk in this period. The April to July 2025 period saw all assets experience elevated volatility simultaneously, showcasing the systemic risk. The March 2026 spike saw **oil volatility reach nearly 100% annualised** while equities rose only modestly, showcasing of isolated supply risk.

A finding that should concern risk managers is that **oil volatility never fully returned to its pre-2025 baseline** even between major events. It remained structurally elevated at 25-35% annualised during the calmer months of late 2025. Markets were pricing in persistent geopolitical risk rather than treating each event as isolated. For businesses dependent on oil this means elevated hedging costs even in periods of apparent calm, and for central banks it means the threat 
of oil driven inflation remained live throughout the period.

### Correlation Matrix

The correlation matrix reveals these four assets were **largely independent of each other** during this period. Oil showed weak negative correlations with GBP/USD (-0.089) and FTSE (-0.169) — directionally logical given the UK is a net oil importer, but weak enough to suggest the FTSE's significant energy sector weighting partially offset the negative impact of higher oil costs on other sectors. The **near zero correlation between oil and the S&P 500 (0.041)** reflects America's unique position as simultaneously the world's largest oil producer and consumer, higher prices benefit energy companies while hurting consumers, roughly cancelling at index level.

The **strongest correlation was FTSE and S&P 500 at 0.243** — both driven by shared global risk sentiment. This has an important implication for UK investors who believe they are diversifying by holding both UK and US equities. During periods of global stress this diversification provides less protection than it appears. **Correlation does not imply causation**, where assets moved together it reflects shared exposure to geopolitical risk rather than any direct causal relationship.

### Event Study

The event study provides the most precise measure of geopolitical shock impact. The **Trump tariff announcements produced the largest systemic shock**, oil -12.6%, FTSE -10.8% and S&P -3.8% simultaneously. What makes this significant is that tariffs are a policy choice rather than an unpredictable event. Markets had warning signs yet still experienced a severe shock, even anticipated geopolitical risks can produce extreme reactions when they materialise at scale. The **US/Israel strikes on Iran produced the largest isolated shock** — a 25.9% oil impact while equities fell less than 2% and GBP/USD barely moved.

The **Iran ceasefire finding is subtle but important** — oil had already been falling 12.4% in the 5 days before the announcement as markets anticipated resolution. By the time the news was public, most of the move had already happened. This is consistent with research on market efficiency — **acting on public geopolitical news is almost always too late**. The OPEC+ output increase producing a positive oil shock is also counterintuitive, more supply should push prices down. The positive reaction suggests markets interpreted increased output as a signal of OPEC+ confidence in sustained demand rather than oversupply.

### Answering the Central Question

**The evidence is clear. Trade and tariff conflicts created systemic risk, simultaneous declines across multiple asset classes with few safe havens. Military conflicts concentrated in energy markets with limited spillover, severe for oil dependent businesses but largely contained for diversified investors.**

This analysis demonstrates that **not all geopolitical shocks are equal**. Trade and tariff conflicts create systemic risk hitting all asset classes 
simultaneously, while military conflicts concentrate in energy markets creating isolated supply shocks. Understanding which type of shock you are facing is the first and most important step in responding appropriately. The distinction between the Trump tariff impact and the US/Iran strikes makes this clear, same period, fundamentally different market dynamics.
